```python
from datetime import datetime
from pyspark.sql import SparkSession

spark = (SparkSession.builder.appName("pyspark-rdd-demo-{}".format(datetime.today()))
        .master("spark://spark-master:7077")      
        .getOrCreate())

spark.sparkContext.getConf().getAll()
```

```python
users = spark.read.format("parquet").load("s3a://warehouse/gold/tiki/users.parquet")
reviews = spark.read.format("parquet").load("s3a://warehouse/gold/tiki/reviews.parquet")
```

```python
users = users.toPandas()
reviews = reviews.toPandas()
```

```python
reviews.info()
```

```python
import re
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Đọc danh sách từ dừng từ tệp văn bản
with open('vietnamese-stopwords.txt', 'r', encoding='utf-8') as f:
    vietnamese_stopwords = set(f.read().splitlines())

# Hàm tiền xử lý văn bản
def preprocess_text(text):
    # Tách từ
    tokens = re.findall(r'\b\w+\b', text.lower())
    # Loại bỏ từ dừng và từ ngắn
    tokens = [word for word in tokens if word not in vietnamese_stopwords and len(word) > 2]
    return ' '.join(tokens)

# Kết hợp nội dung từ title và content
reviews['combined_content'] = reviews['title'].fillna('') + ' ' + reviews['content'].fillna('')

# Áp dụng tiền xử lý cho cột 'combined_content'
reviews['cleaned_combined_content'] = reviews['combined_content'].apply(preprocess_text)

# Kết hợp tất cả các bình luận đã làm sạch thành một chuỗi văn bản duy nhất
cleaned_comments_text = ' '.join(reviews['cleaned_combined_content'].dropna().astype(str))

# Tạo đám mây từ với kích thước lớn hơn
wordcloud = WordCloud(width=1600, height=800, background_color='white', max_words=200, stopwords=None).generate(cleaned_comments_text)

# Tăng độ phân giải khi lưu hình ảnh
plt.figure(figsize=(16, 8))  # Kích thước lớn hơn khi hiển thị
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud of Combined Title and Content')

# Lưu đám mây từ vào tệp hình ảnh (tăng dpi để hình ảnh sắc nét hơn)
plt.savefig('wordcloud.png', format='png', bbox_inches='tight', dpi=300)  # Thêm dpi=300 để tăng độ phân giải

# Loại bỏ hoặc bình luận dòng plt.show() để không hiển thị
# plt.show()
```
